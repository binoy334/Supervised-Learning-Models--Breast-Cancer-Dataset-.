```python
# 1. Loading and Preprocessing
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load the dataset
data = load_breast_cancer()

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
# Checking for Missing Values
# Before proceeding with feature scaling, it’s essential to check for missing values:
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each feature:\n", missing_values[missing_values > 0])
```

Handling missing values is critical because they can lead to biased estimates or errors in model training.

```python
# Checing for duplicates
print("Duplicated values in each feature : \n",df.duplicated().sum())
```

```python
# Checking for Outliers
# Outliers can significantly impact the performance of machine learning models. We can use box plots to visualize potential outliers:
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize outliers using box plots for all features
plt.figure(figsize=(15, 10))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    sns.boxplot(y=df[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
#  Removing outliers based on IQR for a specific feature  'mean radius'
Q1 = df['mean radius'].quantile(0.25)
Q3 = df['mean radius'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
df_cleaned = df[(df['mean radius'] >= lower_bound) & (df['mean radius'] <= upper_bound)]
outlier = df[(df['mean radius'] < lower_bound) & (df['mean radius'] > upper_bound)]
outlier
```

As we can see from the describe function as well as the outlier handling technique performed on the mean radius feautre, there are no outliers or the present outlier  does not affect the dataset significantly.

```python
# Checking for Skewness
# Check skewness of features
skewness = df_cleaned.drop('target', axis=1).skew()
print("Skewness of features:\n", skewness)

# Visualize skewness using histograms
plt.figure(figsize=(15, 10))
for i, column in enumerate(df_cleaned.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    sns.histplot(df_cleaned[column], kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import boxcox

# Apply Box-Cox transformation to positively skewed features (example: 'mean area')
df_cleaned['mean area'], _ = boxcox(df_cleaned['mean area'] + 1)  # Adding 1 to avoid log(0)
```

```python
# Feature Scaling
# After handling outliers and skewness, scale the features:A
from sklearn.preprocessing import StandardScaler

# Separate features and target variable after cleaning
X = df_cleaned.drop('target', axis=1)
y = df_cleaned['target']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

In the preprocessing phase of the breast cancer dataset analysis, several essential steps were undertaken to ensure that the data was clean, well-structured, and suitable for machine learning models. 
First, the dataset was loaded from the scikit-learn library and converted into a pandas DataFrame for easier manipulation.
Checking for missing values was crucial, as any absence of data could lead to biased estimates and negatively impact model performance; fortunately, this dataset contained no missing values. 
Next, outliers were identified using box plots, as they can skew results and lead to inaccurate predictions; if present, outliers would need to be removed or transformed to enhance model robustness. Additionally, skewness in feature distributions was assessed since many machine learning algorithms assume normally distributed data; addressing significant skewness through transformations such as log or square root could stabilize variance and improve performance.
Feature scaling was then performed using StandardScaler to standardize the features by removing the mean and scaling them to unit variance. This step is particularly important for algorithms sensitive to feature scales, such as Support Vector Machines (SVM) and k-Nearest Neighbors (k-NN), as unscaled features could lead to biased model training. 
Finally, the dataset was split into training and testing sets (80% for training and 20% for testing) to effectively evaluate model performance; this separation is critical for assessing how well the model generalizes to unseen data and helps prevent overfitting. Overall, these preprocessing steps are vital for preparing the breast cancer dataset, ensuring high-quality data that enhances the performance and reliability of machine learning algorithms.

## 2. Classification Algorithm Implementation (5 marks)

```python
# Splitting the Dataset
# Now that we have preprocessed our data, we can split it into training and testing sets:
from sklearn.model_selection import train_test_split

# Split data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

### Logistic Regression
How It Works: Logistic Regression is a statistical model that uses a logistic function to model a binary dependent variable. It estimates the probability that a given input point belongs to a certain class. The output is transformed using the logistic function to produce values between 0 and 1, which can be interpreted as probabilities.

Suitability for Dataset: Logistic Regression is suitable for the breast cancer dataset because it is inherently a binary classification problem (malignant vs benign). It provides interpretable coefficients that indicate the influence of each feature on the likelihood of cancer being malignant, making it easier to understand relationships within the data.

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

### Decision Tree Classifier
How It Works: A Decision Tree Classifier creates a model that predicts the value of a target variable based on several input variables. It does this by splitting the data into subsets based on feature values, creating branches until it reaches leaf nodes that represent class labels. The splits are determined by measures such as Gini impurity or entropy to ensure that each split maximizes the homogeneity of the resulting subsets.

Suitability for Dataset: Decision Trees are suitable for this dataset because they can handle both numerical and categorical data, require little preprocessing, and provide clear visualization of decision-making processes. They are also robust to outliers and can capture non-linear relationships between features.

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

### Random Forest Classifier
How It Works: Random Forest is an ensemble learning method that constructs multiple decision trees during training and outputs the mode of their predictions (for classification tasks). It introduces randomness by selecting a subset of features for each tree, which helps reduce overfitting and improves generalization.

Suitability for Dataset: Random Forest is particularly effective for the breast cancer dataset due to its ability to handle high-dimensional data and its robustness against overfitting. The ensemble approach typically results in higher accuracy compared to individual decision trees, making it suitable for complex datasets like this one.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

### Support Vector Machine (SVM)
How It Works: SVM is a supervised learning algorithm that finds the hyperplane that best separates different classes in high-dimensional space. It works by maximizing the margin between data points of different classes, effectively creating a boundary that minimizes classification errors

Suitability for Dataset: SVM is effective for the breast cancer dataset because it performs well in high-dimensional spaces and can handle cases where classes are not linearly separable through the use of kernel functions. This flexibility allows SVM to capture complex relationships in the data.

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(X_train, y_train)
```

### k-Nearest Neighbors (k-NN)
How It Works: k-NN is a simple, instance-based learning algorithm that classifies new instances based on their proximity to existing instances in the training set. The class of a new instance is determined by majority voting among its k-nearest neighbors in feature space.

Suitability for Dataset: k-NN is suitable for this dataset due to its simplicity and effectiveness with small datasets. It does not assume any underlying distribution of data and can easily adapt to different types of data distributions, making it flexible for various scenarios within the breast cancer dataset.

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

## 3. Model Comparison (2 marks)

```python
# Now we will evaluate each model's performance on the test set:
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the models to evaluate
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "k-NN": KNeighborsClassifier()
}

# Function to evaluate models and return metrics
def evaluate_model(model):
    # Fit the model on training data
    model.fit(X_train, y_train)
    
    # Make predictions on test data
    predictions = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    
    # Generate classification report (includes precision, recall, F1-score)
    report = classification_report(y_test, predictions, target_names=["Malignant", "Benign"])
    
    # Generate confusion matrix
    conf_matrix = confusion_matrix(y_test, predictions)
    
    return accuracy, report, conf_matrix

# Evaluate all models and store results in a dictionary
results = {}
for name, model in models.items():
    accuracy, report, conf_matrix = evaluate_model(model)
    results[name] = {
        "Accuracy": accuracy,
        "Report": report,
        "Confusion Matrix": conf_matrix
    }


import matplotlib.pyplot as plt

# Display results for each model
for name, metrics in results.items():
    print(f"Model: {name}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print("Classification Report:\n", metrics['Report'])
    print("Confusion Matrix:\n", metrics['Confusion Matrix'])
    sns.heatmap(conf_matrix, annot=True)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix for the Breast Cancer Dataset')
    plt.show()
    print("=" * 50)  # Separator for better readability
```

### Which algorithm performed the best and which one performed the worst?

### Best Performing Algorithm
Logistic Regression and Support Vector Machine (SVM) both achieved the highest accuracy of 96.40%.

They also had very similar classification reports, with high precision, recall, and F1-scores for both classes (Malignant and Benign).

### Worst Performing Algorithm
The Decision Tree Classifier performed the worst with an accuracy of 90.09%.

It had lower precision and recall values compared to the other models, particularly for the Malignant class.
